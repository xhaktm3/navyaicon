```python
#contrast 변경

from PIL import Image
from PIL import ImageEnhance

def adjust_contrast(image_path, output_path, factor):
    # 이미지 열기
    image = Image.open(image_path)
    
    # 대비 조정
    enhancer = ImageEnhance.Contrast(image)
    adjusted_image = enhancer.enhance(factor)
    
    # 조정된 이미지 저장
    adjusted_image.save(output_path)

# 이미지 경로와 대비 조정 계수(factor) 설정
image_path = "/home/ubuntu/workspace/datasets/final_dataset/train/EO/image/dji_mavick_mountain_cruise_419.jpg"  # 본인의 이미지 경로로 대체해주세요
output_path = "/home/ubuntu/workspace/visualization/output_image.jpg"  # 조정된 이미지의 저장 경로
contrast_factor = 1.5  # 대비 조정 계수 (1보다 크면 대비 증가, 1보다 작으면 대비 감소)

# 이미지 대비 조정
adjust_contrast(image_path, output_path, contrast_factor)
```

```python
from PIL import Image
from PIL import ImageEnhance
import matplotlib.pyplot as plt

def adjust_contrast(image_path, factor):
    # 이미지 열기
    image = Image.open(image_path)
    
    # 대비 조정
    enhancer = ImageEnhance.Contrast(image)
    adjusted_image = enhancer.enhance(factor)
    
    # 원본 이미지와 조정된 이미지 디스플레이
    fig, axs = plt.subplots(1, 2)
    axs[0].imshow(image)
    axs[0].set_title('Original Image')
    axs[0].axis('off')
    axs[1].imshow(adjusted_image)
    axs[1].set_title('Adjusted Contrast Image')
    axs[1].axis('off')
    plt.show()

# 이미지 경로와 대비 조정 계수(factor) 설정
image_path = "/home/ubuntu/workspace/datasets/final_dataset/train/EO/image/GOPR5847_004_545.jpg" # 본인의 이미지 경로로 대체해주세요
contrast_factor = 1.5  # 대비 조정 계수 (1보다 크면 대비 증가, 1보다 작으면 대비 감소)

# 이미지 대비 조정 및 디스플레이
adjust_contrast(image_path, contrast_factor)
```

```python
from PIL import Image
from PIL import ImageEnhance
import matplotlib.pyplot as plt

def adjust_contrast(image_path, factor):
    # 이미지 열기
    image = Image.open(image_path)
    
    # 대비 조정
    enhancer = ImageEnhance.Contrast(image)
    adjusted_image = enhancer.enhance(factor)
    
    # 원본 이미지와 조정된 이미지 디스플레이
    fig, axs = plt.subplots(1, 2)
    axs[0].imshow(image)
    axs[0].set_title('Original Image')
    axs[0].axis('off')
    axs[0].set_aspect('auto')  # 이미지를 쪽에 꽉 차도록 표시
    axs[1].imshow(adjusted_image)
    axs[1].set_title('Adjusted Contrast Image')
    axs[1].axis('off')
    axs[1].set_aspect('auto')  # 이미지를 쪽에 꽉 차도록 표시
    plt.tight_layout()  # 그림 간격 조정
    plt.show()

# 이미지 경로와 대비 조정 계수(factor) 설정
image_path = "/home/ubuntu/workspace/datasets/final_dataset/test/image/00005.jpg"  # 본인의 이미지 경로로 대체해주세요
contrast_factor = 1.5  # 대비 조정 계수 (1보다 크면 대비 증가, 1보다 작으면 대비 감소)

# 이미지 대비 조정 및 디스플레이
adjust_contrast(image_path, contrast_factor)
```
