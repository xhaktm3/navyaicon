```python
from pylabel import importer
```

```python
#Specify path to the coco.json file
path_to_annotations = "/home/ubuntu/workspace/visualization/augmented_data/train.json"
#Specify the path to the images (if they are in a different folder than the annotations)
path_to_images = "/home/ubuntu/workspace/visualization/augmented_data/image"

#Import the dataset into the pylable schema 
dataset = importer.ImportCoco(path_to_annotations, path_to_images=path_to_images, name="drone")
dataset.df.head(5)
```

```python
dataset.path_to_annotations = "/home/ubuntu/workspace/visualization/augmented_data/labels"
dataset.export.ExportToYoloV5(output_path="/home/ubuntu/workspace/visualization/augmented_data/labels")
```
