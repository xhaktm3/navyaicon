```python
from pylabel import importer
```

```python
#Specify path to the coco.json file
path_to_annotations = "/home/jupyter/dacon/data/sample/annotations/updated_data.json"
#Specify the path to the images (if they are in a different folder than the annotations)
path_to_images = "/home/jupyter/dacon/data/sample/train"

#Import the dataset into the pylable schema 
dataset = importer.ImportCoco(path_to_annotations, path_to_images=path_to_images, name="coco")
dataset.df.head(40)
```

```python
dataset.export.ExportToVoc(output_path="/home/jupyter/dacon/data/sample/annotations/")
```
